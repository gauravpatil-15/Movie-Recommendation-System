```python
movies = movies.merge(credits,on='title')                   # To merge credits data over movies using similar column title
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]              # just keeping these columns
```

```python
movies.isnull().sum()                 #To check null values in the data
```

```python
movies.dropna(inplace=True)           #To drop null values
```

```python
movies['genres'][0]
```

```python
import ast
```

```python
def convert(text):                  #To get the main data like   In genres, fetching "name" keys data in the dictionary 
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):                             #To fetch first 3 actors 
    L = []
    counter = 0
    for i in ast.literal_eval(text):
      if counter != 3:
          L.append(i['name'])
          counter=+1
      else:
          break 
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(text):                                #To fetch the director from crew column
    L = []
    for i in ast.literal_eval(text):
      if i['job'] == 'Director':
        L.append(i['name'])
        break 
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())                 #Converting overview from string to list
```

```python
movies.head()
```

```python
#removing spaces 

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']           # combining all the columns      
```

```python
movies.head()
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))                # converting to string
```

```python
new_df = movies[['movie_id','title','tags']]
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())             #Converting to lowercase
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):                             # See below example
  L = []

  for i in text.split():
    L.append(ps.stem(i))

  return " ".join(L)
```

```python
print(stem('loving'))
print(stem('loved'))
stem('love')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer                     #To convert the data into vetor   . Each movie will be unique
cv = CountVectorizer(max_features=5000,stop_words='english')                    #stop_words attribute removes the stop words from the data
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()                             #Converting in array
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vector
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)            # count the similarities betn the movies from count 0 to 1 if 1, its a same movie
```

```python
similarity
```

```python
new_df[new_df['title'] == 'John Carter'].index[0]     # To fetch the index of the movie from the data
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])
    
    for i in movies_list[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Pirates of the Caribbean: At World's End")
```
